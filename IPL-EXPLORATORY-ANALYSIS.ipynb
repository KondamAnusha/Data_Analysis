```python
# we are going to slice the data frame where were all the rows cost in Rs is greater than 0 should be unique
teams= data[data['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
data['STATUS']= data['Team'].replace(teams,'sold')
data.tail()
```

```python
import matplotlib.pyplot as plyt
```

```python
# How many different types of players participated
types = data['TYPE'].value_counts()
types.reset_index()
```

```python
data['STATUS'].value_counts()
```

```python
# plotting pie chart for knowing the each type of players(Piee chart helps to know relative distribution as well as contribution)
#matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)
plyt.pie(types.values,labels=types.index, autopct='%2.2f%%', shadow=True,startangle=60)
plyt.title('ROLE OF PLAYERS PARTICIPATED', fontsize=23)
plyt.plot()
```

```python
#Players unsold and sold

plyt.figure(figsize=(10,10))
fig=sns.countplot(data['STATUS'],palette=['Green','Red'])
plyt.xlabel('Sold or Unsold')
plyt.ylabel('Number of Players')
plyt.title('Sold Vs Unsold', fontsize=15)
plyt.plot()
```
